```python
#12
import pandas as pd

file_path = 'E:/.kaggle/StateNames.csv'
data = pd.read_csv(file_path)

data_2007 = data[data['Year'] == 2007]

unique_names_2007 = data_2007['Name'].nunique()

most_popular_name_2007 = data_2007.loc[data_2007['Count'].idxmax()]

print(f"Кількість унікальних імен у 2007 році: {unique_names_2007}")
print("Найпопулярніше ім'я у 2007 році:")
print(f"Ім'я: {most_popular_name_2007['Name']}, Стать: {most_popular_name_2007['Gender']}, Count: {most_popular_name_2007['Count']}")
```

```python
#18
import pandas as pd

file_path = 'E:/.kaggle/StateNames.csv'
data = pd.read_csv(file_path)

grouped_data = data.groupby(['Year', 'Gender'])['Count'].sum().unstack()

years_more_girls = (grouped_data['F'] > grouped_data['M']).sum()

print(f"Кількість років, коли дівчаток народжувалось більше, ніж хлопчиків: {years_more_girls}")
```

```python
#19
import pandas as pd
import matplotlib.pyplot as plt

file_path = 'E:/.kaggle/StateNames.csv'
data = pd.read_csv(file_path)

births_by_gender = data.groupby(['Year', 'Gender'])['Count'].sum().unstack()

births_by_gender.rename(columns={'F': 'Girls', 'M': 'Boys'}, inplace=True)

plt.figure(figsize=(12, 6))
plt.plot(births_by_gender.index, births_by_gender['Girls'], label='Girls', color='pink', linewidth=2)
plt.plot(births_by_gender.index, births_by_gender['Boys'], label='Boys', color='blue', linewidth=2)

plt.title('Загальна кількість народжень хлопчиків та дівчаток на рік', fontsize=16)
plt.xlabel('Рік', fontsize=12)
plt.ylabel('Кількість народжень', fontsize=12)
plt.legend(fontsize=12)
plt.grid(alpha=0.3)
plt.tight_layout()

plt.show()
```
